```python
import pandas as pd

df = pd.read_csv('GlobalWeatherRepository.csv')

print("Dataset loaded successfully")
df.shape
```

```python
df['last_updated'] = pd.to_datetime(df['last_updated'])

df['year'] = df['last_updated'].dt.year
df['month'] = df['last_updated'].dt.month
df['date'] = df['last_updated'].dt.date
```

```python
import numpy as np

num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.drop_duplicates(inplace=True)

df['country'] = df['country'].str.title()
df['location_name'] = df['location_name'].str.title()
df['condition_text'] = df['condition_text'].str.lower().str.strip()
```

```python
drop_cols = [
    'temperature_fahrenheit',
    'wind_mph',
    'pressure_in',
    'precip_in'
]

df.drop(columns=drop_cols, inplace=True, errors='ignore')
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scale_cols = [
    'temperature_celsius',
    'humidity',
    'wind_kph',
    'air_quality_PM2.5',
    'air_quality_PM10'
]

df[scale_cols] = scaler.fit_transform(df[scale_cols])
```

```python
monthly_df = df.groupby(
    ['country', 'location_name', 'year', 'month']
).agg({
    'temperature_celsius': 'mean',
    'humidity': 'mean',
    'air_quality_PM2.5': 'mean',
    'air_quality_PM10': 'mean'
}).reset_index()
```

```python
df.to_csv("cleaned_weather_data.csv", index=False)
monthly_df.to_csv("monthly_weather_data.csv", index=False)

print("Preprocessed datasets saved successfully")
```
