```python
import pandas as pd

df = pd.read_csv("../data/raw/GlobalWeatherRepository.csv")
df.shape
```

```python
df.info()
df.describe()
```

```python
df.duplicated().sum()

```

```python
df['last_updated'] = pd.to_datetime(df['last_updated'])
```

```python
df['year'] = df['last_updated'].dt.year
df['month'] = df['last_updated'].dt.month
df['day'] = df['last_updated'].dt.day
df['year_month'] = df['last_updated'].dt.to_period('M')
```

```python
print("Duplicates:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
df[['wind_kph', 'gust_kph', 'pressure_mb']].describe()
```

```python
import numpy as np

df.loc[df['wind_kph'] > 200, 'wind_kph'] = np.nan
df.loc[df['gust_kph'] > 250, 'gust_kph'] = np.nan
df.loc[df['pressure_mb'] > 1100, 'pressure_mb'] = np.nan
df.loc[df['pressure_mb'] < 900, 'pressure_mb'] = np.nan
```

```python
air_cols = [
    'air_quality_Carbon_Monoxide',
    'air_quality_Ozone',
    'air_quality_Nitrogen_dioxide',
    'air_quality_Sulphur_dioxide',
    'air_quality_PM2.5',
    'air_quality_PM10'
]

for col in air_cols:
    df.loc[df[col] < 0, col] = np.nan
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
df.isnull().sum()
```

```python
selected_columns = [
    'country',
    'location_name',
    'latitude',
    'longitude',
    'last_updated',
    'year',
    'month',
    'temperature_celsius',
    'feels_like_celsius',
    'humidity',
    'wind_kph',
    'gust_kph',
    'pressure_mb',
    'precip_mm',
    'uv_index',
    'cloud',
    'visibility_km',
    'air_quality_PM2.5',
    'air_quality_PM10'
]

df = df[selected_columns]
```

```python
print(df.columns)

```

```python
df['last_updated'] = pd.to_datetime(df['last_updated'])
df['year_month'] = df['last_updated'].dt.to_period('M')
```

```python
df['last_updated'] = pd.to_datetime(df['last_updated'])

df['year'] = df['last_updated'].dt.year
df['month'] = df['last_updated'].dt.month
df['year_month'] = df['last_updated'].dt.to_period('M')
```

```python
df[['last_updated','year_month']].head()
```

```python
monthly_avg = df.groupby(['country','year_month']).mean(numeric_only=True).reset_index()
```

```python
df.columns
```

```python
df.to_csv("../data/processed/weather_cleaned_daily.csv", index=False)
monthly_avg.to_csv("../data/processed/weather_cleaned_monthly.csv", index=False)
```
