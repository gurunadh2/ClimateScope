```python
import pandas as pd

# Load the daily cleaned dataset from your processed folder
df = pd.read_csv('../data/processed/weather_cleaned_daily.csv')

# Give it a quick check
display(df.head())
```

```python
# 1. Overall Statistical Summary (Distributions)
print("--- Statistical Summary ---")
display(df.describe())

# 2. Correlation Matrix
weather_vars = [
    'temperature_celsius',
    'humidity',
    'wind_kph',
    'precip_mm'           # matches your CSV header
]

print("\n--- Correlation Matrix ---")
display(df[weather_vars].corr())
```

```python
# 1. Identify Extreme Weather Events
print("--- Extreme Weather Events ---")
hottest_record   = df.loc[df['temperature_celsius'].idxmax()]
windiest_record  = df.loc[df['wind_kph'].idxmax()]

print(f"Hottest recorded temperature: {hottest_record['temperature_celsius']}°C in "
      f"{hottest_record['country']}")
print(f"Highest wind speed: {windiest_record['wind_kph']} kph in "
      f"{windiest_record['country']}")

# 2. Regional Comparisons (Average temperature by country)
print("\n--- Regional Temperature Comparison ---")
regional_avg = (
    df.groupby('country')['temperature_celsius']
      .mean()
      .sort_values(ascending=False)
)
display(regional_avg.head(10))   # top‑10 hottest countries on average
```
